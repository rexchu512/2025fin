```python
import numpy as np
import scipy.stats as si
import matplotlib.pyplot as plt

# 初始化
T = []
C = []

# 第一次計算
d1 = (np.log(100/100)+(0.1+0.5**2/2)*1)/(0.5*np.sqrt(1))
d2 = (np.log(100/100)+(0.1+0.5**2/2)*1)/(0.5*np.sqrt(1))-0.5*np.sqrt(1)
call = 100*si.norm.cdf(d1,0.0,1.0)-100*np.exp(-0.1*1)*si.norm.cdf(d2,0.0,1.0)
T.append(1)
C.append(call)

# 第二次計算
d1 = (np.log(100/100)+(0.1+0.5**2/2)*0.5)/(0.5*np.sqrt(0.5))
d2 = (np.log(100/100)+(0.1+0.5**2/2)*0.5)/(0.5*np.sqrt(0.5))-0.5*np.sqrt(0.5)
call = 100*si.norm.cdf(d1,0.0,1.0)-100*np.exp(-0.1*0.5)*si.norm.cdf(d2,0.0,1.0)
T.append(0.5)
C.append(call)

# 第三次計算
d1 = (np.log(100/100)+(0.1+0.5**2/2)*0.3)/(0.5*np.sqrt(0.3))
d2 = (np.log(100/100)+(0.1+0.5**2/2)*0.3)/(0.5*np.sqrt(0.3))-0.5*np.sqrt(0.3)
call = 100*si.norm.cdf(d1,0.0,1.0)-100*np.exp(-0.1*0.3)*si.norm.cdf(d2,0.0,1.0)
T.append(0.3)
C.append(call)

# 第四次計算
d1 = (np.log(100/100)+(0.1+0.5**2/2)*0.1)/(0.5*np.sqrt(0.1))
d2 = (np.log(100/100)+(0.1+0.5**2/2)*0.1)/(0.5*np.sqrt(0.1))-0.5*np.sqrt(0.1)
call = 100*si.norm.cdf(d1,0.0,1.0)-100*np.exp(-0.1*0.1)*si.norm.cdf(d2,0.0,1.0)
T.append(0.1)
C.append(call)

# 繪圖
plt.plot(T, C)
plt.xlabel('Time to Maturity (T)')
plt.ylabel('Call Option Price')
plt.title('Call Option Price vs Time to Maturity')
plt.grid(True)
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(T,C)
plt.xlim(-0.05,1.2)
ax.invert_xaxis()
plt.show()
```
