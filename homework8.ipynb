```python
import numpy as np
import scipy.stats as si

# 固定參數
S = 90        # 資產價格
K = 100       # 履約價
r = 0.01      # 無風險利率
sigma = 0.3   # 波動度

T_list = [1, 0.9, 0.5, 0.3, 0.1]  # 不同到期日

# 計算並印出每個 T 對應的買權價格
for T in T_list:
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * si.norm.cdf(d1) - K * np.exp(-r * T) * si.norm.cdf(d2)
    print(f"T = {T}, Call Price = {round(call_price, 4)}")
```

```python
import numpy as np
import scipy.stats as si
import matplotlib.pyplot as plt

# 固定參數
S = 90
K = 100
r = 0.01
sigma = 0.3

# 設定 T 的區間（例如從 1 到 0.1，間隔 0.01）
T_values = np.arange(0.1, 1.01, 0.01)[::-1]  # 反過來畫比較直觀
call_prices = []

# 計算每個 T 對應的 call 價格
for T in T_values:
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call = S * si.norm.cdf(d1) - K * np.exp(-r * T) * si.norm.cdf(d2)
    call_prices.append(call)

# 畫圖
plt.figure(figsize=(8, 5))
plt.plot(T_values, call_prices, marker='o')
plt.xlabel('Time to Maturity (T)')
plt.ylabel('European Call Option Price')
plt.title('Call Option Price vs. Time to Maturity')
plt.grid(True)
plt.show()
```
